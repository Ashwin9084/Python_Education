## Scikit-learn Design

Primarily three types of objects
1. Estimators - It estimates some parameter on a dataset . eg. imputer it has a fit and a transform. method.
fit method - fits the dataset and calculates internal method
2. Transformers - transfprm method takes the input and returns  oytput based on the learnings from fit(). It also has a convinience function called fit_transfrom() which fits and then transforms
3. Predictors- LinearRegressionm model is an example of predictor.fit() and predict() are two common functions . it also gives score() function which will evalutate the predictors.

## Feature scaling 

Primaralily two types of feature scalling methods
1. Min-max scaling (Normalisation)
  (value - min )/(max- min)
  sklearn provides a class called   MinMaxScaler for this
2. standardization
  (value - mean )/(std)

##Creating a pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
                        ('imputer', SimpleImputer(strategy = 'median')),
                        # .... add as many as you want
                        ('std_scaler', StandardScaler()),
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
```

```python
housing_num_tr.shape
```

## Selecting a desired model for dragon real estate

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
# model = DecisionTreeRegressor()
# model = LinearRegression()
model.fit(housing_num_tr,housing_labels)
```

```python
some_data = housing.iloc[:5]
```

```python
some_labels = housing_labels.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

## Evaluting the model

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
mse= mean_squared_error(housing_labels , housing_predictions )
rmse = np.sqrt(mse)
rmse
```

## using better evaluations techniques - cross validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,housing_num_tr,housing_labels, scoring='neg_mean_squared_error')
rmse_scores= np.sqrt(- scores)
rmse_scores
```

```python
def print_scores(scores):
  print('scores :', scores)
  print('Mean :', scores.mean())
  print('Standard deviation :', scores.std())
print_scores(rmse_scores)
```

## Saving the model

```python
from joblib import dump,load
dump(model, 'Dragon.joblib')
```

## Testing the model

```python
X_test = strat_test_set.drop('MEDV' , axis= 1)
Y_test = strat_test_set['MEDV'].copy()
X_test_prepared = my_pipeline.transform(X_test)
final_predictions = model.predict(X_test_prepared)
final_mse=mean_squared_error(Y_test,final_predictions)
final_rmse=np.sqrt(final_mse)
final_rmse
```
